```python
'''
7. Crea un nuevo dataset con los pokémon de tipo Water y Fire como primer tipo
'''

pk_water = pokemon[(pokemon['Type 1'] == 'Water')]
pk_water.reset_index(drop=True, inplace=True)
pk_water

pk_fire = pokemon[(pokemon['Type 1'] == 'Fire')]
pk_fire.reset_index(drop=True, inplace=True)
pk_fire

pk_filtrado = pk_water.append(pk_fire)
print(pk_filtrado)

print(len(pk_filtrado))
```

```python
'''
8. Crea un nuevo dataset con los pokemon Legendary
'''

pk_legendary = pokemon[(pokemon['Legendary'] == True)]
pk_legendary.reset_index(drop=True, inplace=True)
pk_legendary
```

```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']
precio = []

for elt in compra:
    precio.append(articulos [elt])
print(precio)

# A. Una función que calcule el precio total de la compra

reduce(lambda x, y: x + y, precio)

# B. Una función que calcule el precio total de la compra con un descuento del 10%

reduce(lambda x, y: x + y, precio)*0.90

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
    

precio_total = reduce(lambda x, y: x + y, precio)
if precio_total > 100:
    precio_final = precio_total*0.9
else: 
    precio_final = precio_total
print(precio_final)
     

# D. Una función que calcule el precio total aplicando el IVA (21%)

base = round(precio_total/1.21, 2)
cuota = round(base*0.21, 2)
print("Base: "+str(base)+"\nIVA (21%): " + str(cuota)+"\n Total factura: "+str(precio_total))


# E.lista los artículos cuyo precio es superior a 20€

list(filter(lambda x: x[1] > 20, articulos.items()))
```
